### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head(-1)
```

#1. Análise do faturamento


```python
#calculando o faturamento total de cada loja
faturamento_loja1 = loja['Preço'].sum()
print(f'O faturamento total da loja 1 é de: R${faturamento_loja1:.2f}')
faturamento_loja2 = loja2['Preço'].sum()
print(f'O faturamento total da loja 2 é de: R${faturamento_loja2:.2f}')
faturamento_loja3 = loja3['Preço'].sum()
print(f'O faturamento total da loja 3 é de: R${faturamento_loja3:.2f}')
faturamento_loja4 = loja4['Preço'].sum()
print(f'O faturamento total da loja 4 é de: R${faturamento_loja4:.2f}')
```

```python
#configurar exibição do grafico
faturamento_total = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
cores = ['#3882EB', '#6038EB', '#384AEB', '#A0A7E4']
total_geral = sum(faturamento_total)

plt.figure(figsize=(10, 6))
plt.bar(lojas, faturamento_total, color=cores)

plt.xlabel('Lojas')
plt.ylabel('Faturamento Total (BRL)')
plt.title('Faturamento por Loja')


for i, valor in enumerate(faturamento_total):
    plt.text(i, valor + 10000, f'R${valor:,.0f}', ha='center', va='bottom', fontsize = 10)

plt.figtext(0.5, -0.05, f"Total Geral de Faturamento: R${total_geral:,.2f}", ha="center", fontsize=12, weight='bold', color = 'black')


plt.show()
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
#mostrar produtos por categoria de cada loja
print("Loja 1 - Produtos por Categoria")
print(loja['Categoria do Produto'].value_counts())
print("\n")

print("Loja 2 - Produtos por Categoria")
print(loja2['Categoria do Produto'].value_counts())
print("\n")

print("Loja 3 - Produtos por Categoria")
print(loja3['Categoria do Produto'].value_counts())
print("\n")

print("Loja 4 - Produtos por Categoria")
print(loja4['Categoria do Produto'].value_counts())
```

```python
#analisando os dados de vendas por categoria

for df in [loja, loja2, loja3, loja4]:
    df.columns = df.columns.str.strip()

categoria_loja = loja['Categoria do Produto'].value_counts()
categoria_loja2 = loja2['Categoria do Produto'].value_counts()
categoria_loja3 = loja3['Categoria do Produto'].value_counts()
categoria_loja4 = loja4['Categoria do Produto'].value_counts()


df_categorias = pd.DataFrame({
    'Loja 1': categoria_loja,
    'Loja 2': categoria_loja2,
    'Loja 3': categoria_loja3,
    'Loja 4': categoria_loja4
}).fillna(0)

cores = ['#3882EB', '#6038EB', '#384AEB', '#A0A7E4']

#volume de vendas
df_categorias.plot(kind='bar', figsize=(10, 6), color=cores)
plt.xlabel('Categoria do Produto')
plt.ylabel('Quantidade Vendida')
plt.title('Quantidade Vendida por Categoria em Cada Loja')
plt.legend(title='Lojas')
plt.xticks(rotation=45)


plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
#calcular media de avaliação de cada loja
media_loja = loja['Avaliação da compra'].mean()
print(f'A média de avaliação da loja 1 é: {media_loja:.2f}')
media_loja2 = loja2['Avaliação da compra'].mean()
print(f'A média de avaliação da loja 2 é: {media_loja2:.2f}')
media_loja3 = loja3['Avaliação da compra'].mean()
print(f'A média de avaliação da loja 3 é: {media_loja3:.2f}')
media_loja4 = loja4['Avaliação da compra'].mean()
print(f'A média de avaliação da loja 4 é: {media_loja4:.2f}')
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias = [media_loja, media_loja2, media_loja3, media_loja4]
cores = ['#3882EB', '#6038EB', '#384AEB', '#A0A7E4']

x = range(len(lojas))
plt.figure(figsize=(9, 5))

#marcando pontos
plt.scatter(x, medias, color=cores)

#adicionando valores em cima dos pontos
for i, nota in enumerate(medias):
    plt.text(i, nota + 0.01, f'{nota:.2f}', ha='center', va='bottom', fontsize=10)

#configurando o grafico
plt.title('Média de Avaliação por Loja', fontsize = 14)
plt.xlabel('Lojas')
plt.ylabel('Média de Avaliação')
plt.xticks(x, lojas)
plt.ylim(3.8, 4.2)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('media_avaliacao.png', dpi=300, bbox_inches='tight')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
#mostrando quais itens sao mais e menos vendidos de loja por loja
produtos_mais_vendidos1 = loja['Produto'].value_counts().head(5).reset_index()
produtos_mais_vendidos1.columns = ['Produto', 'Quantidade Vendida']

produtos_menos_vendidos1 = loja['Produto'].value_counts().tail(5).reset_index()
produtos_menos_vendidos1.columns = ['Produto', 'Quantidade Vendida']

produtos_mais_vendidos2 = loja2['Produto'].value_counts().head(5).reset_index()
produtos_mais_vendidos2.columns = ['Produto', 'Quantidade Vendida']

produtos_menos_vendidos2 = loja2['Produto'].value_counts().tail(5).reset_index()
produtos_menos_vendidos2.columns = ['Produto', 'Quantidade Vendida']

produtos_mais_vendidos3 = loja3['Produto'].value_counts().head(5).reset_index()
produtos_mais_vendidos3.columns = ['Produto', 'Quantidade Vendida']

produtos_menos_vendidos3 = loja3['Produto'].value_counts().tail(5).reset_index()
produtos_menos_vendidos3.columns = ['Produto', 'Quantidade Vendida']

produtos_mais_vendidos4 = loja4['Produto'].value_counts().head(5).reset_index()
produtos_mais_vendidos4.columns = ['Produto', 'Quantidade Vendida']

produtos_menos_vendidos4 = loja4['Produto'].value_counts().tail(5).reset_index()
produtos_menos_vendidos4.columns = ['Produto', 'Quantidade Vendida']

print('Produtos mais vendidos da loja 1:\n', produtos_mais_vendidos1, '\n')
print('Produtos menos vendidos da loja 1:\n', produtos_menos_vendidos1, '\n')

print('Produtos mais vendidos da loja 2:\n', produtos_mais_vendidos2, '\n')
print('Produtos menos vendidos da loja 2:\n', produtos_menos_vendidos2, '\n')

print('Produtos mais vendidos da loja 3:\n', produtos_mais_vendidos3, '\n')
print('Produtos menos vendidos da loja 3:\n', produtos_menos_vendidos3, '\n')

print('Produtos mais vendidos da loja 4:\n', produtos_mais_vendidos4, '\n')
print('Produtos menos vendidos da loja 4:\n', produtos_menos_vendidos4, '\n')
```

```python
loja = pd.DataFrame(loja)
loja_2 = pd.DataFrame(loja2)
loja_3 = pd.DataFrame(loja3)
loja_4 = pd.DataFrame(loja4)

#Adicionar coluna identificando a loja
loja['Loja'] = 'Loja '
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

#Juntar todos os dados
df_unificado = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)


#Carregar e unir os dados de todas as lojas
lojas = [loja, loja2, loja3, loja4]
df_unificado = pd.concat(lojas, ignore_index=True)

#Agrupar e somar as quantidades por produto
vendas_por_produto = df_unificado.groupby('Produto')['Quantidade de parcelas'].sum().reset_index()

#Identificar os 5 mais e 5 menos vendidos
cinco_mais = vendas_por_produto.nlargest(5, 'Quantidade de parcelas')
cinco_menos = vendas_por_produto.nsmallest(5, 'Quantidade de parcelas')

#Juntar os resultados
resultados = pd.concat([cinco_mais, cinco_menos])
resultados['Tipo'] = ['Mais Vendido']*5 + ['Menos Vendido']*5

# Gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.barh(
    resultados['Produto'],
    resultados['Quantidade de parcelas'],
    color=['green']*5 + ['red']*5,  # Verde para mais vendidos, vermelho para menos
    alpha=0.7
)

# Adicionar valores nas barras
for bar in bars:
    width = bar.get_width()
    plt.text(width + (max(resultados['Quantidade de parcelas']))*0.01,  # Posição do texto
        bar.get_y() + bar.get_height()/2,  # Centraliza verticalmente
        f'{int(width)}',  # Formata sem decimais
        va='center',
        ha='left',
        fontsize=10
    )

# Personalizando o grafico
plt.title('Produtos Mais e Menos vendidos (Todas as Lojas)', fontsize=14)
plt.xlabel('Quantidade Total Vendida', fontsize=12)
plt.ylabel('Produto', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.3)

# Legenda personalizada
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='green', alpha=0.7, label='5 Mais Vendidos'),
    Patch(facecolor='red', alpha=0.7, label='5 Menos Vendidos')
]
plt.legend(handles=legend_elements, loc='upper right')

plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
frete_medio_loja = loja["Frete"].mean()
frete_medio_loja2 = loja2["Frete"].mean()
frete_medio_loja3 = loja3["Frete"].mean()
frete_medio_loja4 = loja4["Frete"].mean()


print("Frete Médio por Loja:")
print(f"Loja 1: R$ {frete_medio_loja:.2f}")
print(f"Loja 2: R$ {frete_medio_loja2:.2f}")
print(f"Loja 3: R$ {frete_medio_loja3:.2f}")
print(f"Loja 4: R$ {frete_medio_loja4:.2f}")
```

```python
media_frete = [frete_medio_loja, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]
cores = ['#3882EB', '#6038EB', '#384AEB', '#A0A7E4']
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

#grafico de pizza
plt.figure(figsize=(10, 6))
plt.pie(media_frete, labels=lojas, colors=cores, autopct='%1.1f%%', startangle=90)

# Adiciona um título ao gráfico
plt.title('Participação do Frete Médio por Loja')

# Garante que o gráfico seja um círculo perfeito
plt.axis('equal')

# Exibe o gráfico
plt.tight_layout()
plt.show()
```
